```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
import numpy as np

data = pd.read_csv("heart_rate_spo2_dataset_with_outliers.csv")
data.head()
```

```python
plt.scatter(data.iloc[:,0],data.iloc[:,1])
```

As you can see the graph there are some outliers in the data. we need to find out that data points.

```python
clf = IsolationForest(contamination=0.2)
clf.fit(data)
```

```python
preds = clf.predict(data)
```

```python
index = np.where(preds < 0)
index
```

```python
x = data.values
```

```python
X = data.values
plt.scatter(data.iloc[:,0],data.iloc[:,1])
plt.scatter(x[index,0],x[index,1],edgecolors='r')
```
